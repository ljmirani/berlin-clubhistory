```python
import re
from PIL import Image
import pytesseract

clubs = []

def parse_text_to_dict(text):

    
    # Teile den Text in Zeilen
    lines = text.split('\n')
    
    # Initialisiere Variablen für den aktuellen Clubnamen und die Jahre
    current_name = None
    current_years = None
    
    # Regex-Muster für Jahre
    year_pattern = re.compile(r'\d{4} - \d{4}')
    
    for line in lines:
        # Entferne führende und nachfolgende Leerzeichen
        line = line.strip()
        
        # Suche nach dem Jahrmuster in der Zeile
        year_match = year_pattern.search(line)
        
        if year_match:
            # Wenn ein Jahrmuster gefunden wird, speichere es
            current_years = year_match.group(0)
        elif 'Musikrichtungen & Subkulturen' in line:
            # Wenn die Zeile 'Musikrichtungen & Subkulturen' enthält, ignoriere sie
            continue
        elif line:
            # Wenn die Zeile weder Jahre noch 'Musikrichtungen & Subkulturen' enthält, speichere den Namen
            if current_name and current_years:
                # Wenn wir bereits einen Namen und Jahre haben, speichern wir das Dictionary
                clubs.append({'Name': current_name, 'Jahre': current_years})
                current_name = None
                current_years = None
            current_name = line
            
    # Füge das letzte Club-Dictionary hinzu, falls noch vorhanden
    if current_name and current_years:
        clubs.append({'Name': current_name, 'Jahre': current_years})
    
    return clubs

```

```python
numbers = [i for i in range(4, 10)]
print(numbers)
```

```python

# OCR durchführen
for number in numbers:
    image = Image.open(f'screenshot{number}.png')
    text = pytesseract.image_to_string(image)
# Text parsen und in eine Liste von Dictionaries umwandeln
    clubs = parse_text_to_dict(text)
# Ergebnis ausgeben
print(clubs)    



# Parsing-Funktion
```

```python
import pdfplumber
import pandas as pd

def pdf_to_dataframe(pdf_path):
    # Open the PDF file
    pdf = pdfplumber.open(pdf_path)

    # Initialize a list to hold the extracted data
    data = []

    # Iterate through each page
    for page in pdf.pages:
        # Extract tables from the current page
        tables = page.extract_tables()
        for table in tables:
            # Append the table data to the list
            data.extend(table)

    # Close the PDF file
    pdf.close()

    # Convert the data to a DataFrame
    if data:
        df = pd.DataFrame(data[1:], columns=data[0])
    else:
        df = pd.DataFrame()

    return df

# Path to your PDF file
pdf_path = 'screenshot6.pdf'

# Convert PDF to DataFrame
df = pdf_to_dataframe(pdf_path)

# Display the DataFrame
print(df)
```
